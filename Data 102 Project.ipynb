# Prediction with GLMs and Nonparametric Methods

## Methods

**Research Question 1:** Can we predict a candidate's voting outcome during the 2018 primary elections based on their financial data and endorsements?

We believe that money and endorsements play an essential role during the elections. Therefore, for this research question, we aim to develop a model that will make a prediction of whether a candidate would win or lose during the primary elections based on a candidate's campaign finance data and endorsements. Here, we will observe the data from the 2018 primary elections. 

After analyzing the datasets and performing feature selection, we have decided to look at various financial categories, such as a candidate's disbursements, contributions, loans, refunds, and expenditures throughout the 2018 primary elections. We believe these features would be beneficial for our prediction since we noticed a trend during our EDA that candidates who would spend more on their campaigns would have a higher chance of securing a victory during the primary election. Furthermore, we also decided to use the information on candidate endorsements for our predictions since we believe that endorsements can influence the candidates' network and public image.

In order for us to make the prediction, first, we would develop a GLM. For this part, we will be performing logistic regression to predict whether a candidate would win or lose the primary. Therefore, our likelihood model for the GLM would be a Bernoulli (or Binomial) distribution with the logit function and sigmoid function as the link and inverse link functions. 

Furthermore, we would also invoke nonparametric methods to make our prediction. For this part, we will be using random forests. We decided to use this method since we are working with numerical data,therefore, random forests will be a great method for our prediction.

To evaluate each model's performance, we will analyze and compare each model's prediction accuracy and the Area Under the Curve (AUC) score of the ROC curve of each model. The model with the highest accuracy score and AUC-ROC score will be considered the best model.

### Setup and Datasets

```python
# Python Libraries

import arviz as az
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split
from statsmodels.formula.api import glm
import theano

try:
    from pymc3 import *
    import pymc3 as pm
except:
    ! pip install pymc3
    from pymc3 import *
    import pymc3 as pm
```

```python
# Candidate Finances (2018 Primary)
finances = pd.read_csv('finances.csv', index_col = 0)
```

```python
# Democrat Candidate Endorsements (2018 Primary)
endorsements = pd.read_csv('democrat_endorsements.csv', index_col = 0)
```

```python
# Cleanup Candidate Names for endorsements
temp = endorsements['Candidate']
temp = temp.str.replace(r'\s\([^()]*\)', "", regex=True)
temp = temp.str.replace(r'\s.+\.', "", regex=True)
temp = temp.str.replace(r'\s\".+\"', "", regex=True)
temp = temp.str.upper()
endorsements['Candidate'] = temp
endorsement_columns = ['Candidate', 'Veteran?', 'LGBTQ?', 'Elected Official?', 'Self-Funder?', 'STEM?','Obama Alum?', 'Party Support?', 'Emily Endorsed?','Guns Sense Candidate?', 'Biden Endorsed?', 'Warren Endorsed? ','Sanders Endorsed?', 'Our Revolution Endorsed?','Justice Dems Endorsed?', 'PCCC Endorsed?', 'Indivisible Endorsed?','WFP Endorsed?', 'VoteVets Endorsed?', 'No Labels Support?']

# Combined Data of Candidate Finances and Endorsements
finance_endorsements = pd.merge(finances, endorsements[endorsement_columns], left_on='Candidate', right_on='Candidate')
finance_endorsements.columns = finance_endorsements.columns.str.replace(' ','_')
finance_endorsements.columns = finance_endorsements.columns.str.replace('?','', regex=False)
```

### Helper Functions

```python
def normalize(data):
    columns = data.columns
    return pd.DataFrame(preprocessing.normalize(data), columns = columns)
```

```python
def select_features(X, Y, X_test, k = 10):
    columns = X.columns
    features = SelectKBest(score_func=f_regression, k = k)
    features.fit(X, Y)
    cols = features.get_support(indices=True)
    return X.iloc[:, cols], X_test.iloc[:, cols]
```

```python
def accuracy(Y, Y_hat):
    return np.count_nonzero(Y == Y_hat) / len(Y)
```

```python
def formula(features, y_label):
    formula = y_label + ' ~ '
    for i in np.arange(len(features)):
        if i == 0:
            formula += features[i]
        else:
            formula += ' + ' + features[i]
    return formula
```

### GLM (Frequentist)

(1) Setup Training Data

```python
pd.options.mode.chained_assignment = None
np.seterr(invalid='ignore')

# Split Dataset to Training and Testing Dataset
train, test = train_test_split(finance_endorsements, test_size = 0.33, random_state = 10)

# Y Training
Y_train = [1 if y == 'Yes' else 0 for y in train['Won_Primary']]

# X Training
train.replace([np.inf, -np.inf], np.nan, inplace=True)
train.dropna(inplace=True, axis = 1)
X_train = train.select_dtypes([np.number])

# Y test
Y_test = [1 if y == 'Yes' else 0 for y in test['Won_Primary']]

# X test
X_test = test[X_train.columns]

# Feature Selection
X_train, X_test = select_features(X_train, Y_train, X_test, k = 35)
```

(2) Fit GLM Model

```python
# Fit GLM Model Using Training Data
glm = sm.GLM(Y_train, X_train, family = sm.families.Binomial()).fit()
```

(3) Evaluate GLM Model

```python
# Prediction on Training Set
Y_train_pred = glm.predict(X_train) > 0.5

# Training Accuracy Score
training_acc = accuracy(Y_train, Y_train_pred)

# Training ROC-AUC Score
training_auc = roc_auc_score(Y_train, Y_train_pred)

# Prediction on Test Set
Y_test_pred = glm.predict(X_test) > 0.5

# Test Accuracy Score
test_acc = accuracy(Y_test, Y_test_pred)

# Test ROC-AUC Score
test_auc = roc_auc_score(Y_test, Y_test_pred)
```

```python
print("Accuracy Scores (Frequentist GLM)")
print("-------------------------------------------")

print(f"Training Accuracy Score: {training_acc}")

print(f"Test Accuracy Score: {test_acc}")

print()

print("ROC-AUC Scores (Frequentist GLM)")

print("-------------------------------------------")

print(f"Training ROC-AUC Score: {training_auc}")

print(f"Test ROC-AUC Score: {test_auc}")
```

(1) Bayesian Regression

```python
bayes_shared = theano.shared(X_train.copy()[columns].values)

with pm.Model() as logistic_model_pred:
    pm.glm.GLM(x=bayes_shared, labels=columns, y = np.array(Y_train), family = pm.glm.families.Binomial())

with logistic_model_pred:
    pred_trace =  pm.sample(1000, tune = 2000, cores = 1, init = 'adapt_diag', target_accept=0.95, return_inferencedata=True)
```

(2) Evaluate GLM Model: Posteror Predictive Checks

```python
# Perform Posterior Predictive Checks
bayes_shared.set_value(X_train.copy()[columns].values)

# Training Dataset Results
with logistic_model_pred:
    logistic_ppc = pm.sample_posterior_predictive(pred_trace, model = logistic_model_pred)
    logistic_ppc['y'] = logistic_ppc['y'] + 0.0
    
Y_train_ppc = np.mean(logistic_ppc['y'], axis = 0) > 0.5

train_ppc_acc = accuracy(Y_train, Y_train_ppc)
train_ppc_auc = roc_auc_score(Y_train, Y_train_ppc)
```

```python
bayes_shared.set_value(X_test.copy()[columns].values)

logistic_ppc_test = pm.sample_posterior_predictive(pred_trace, model = logistic_model_pred)

Y_test_ppc = np.mean(logistic_ppc_test['y'], axis = 0) > 0.5

test_ppc_acc = accuracy(Y_test, Y_test_ppc)
test_ppc_auc = roc_auc_score(Y_test, Y_test_ppc)
```

```python
print("Accuracy Scores (Bayesian GLM)")
print("-------------------------------------------")

print(f"Training Accuracy Score: {train_ppc_acc}")

print(f"Test Accuracy Score: {test_ppc_acc}")

print()

print("ROC-AUC Scores (Bayesian GLM)")

print("-------------------------------------------")

print(f"Training ROC-AUC Score: {train_ppc_auc}")

print(f"Test ROC-AUC Score: {test_ppc_auc}")
```

# **Method: Causal Inference** 
**From this part onwards, it will be focusing on implementation of causal inference.**

**The question that we will be trying to answer using this method is:**

*   Does the amount of net contribution made by a candidate have a causal effect on winning the primary election?

**Treatment**: People who have a “Net_Contribution” over $0.

**Control**: People have a “Net_Contribution” of $0.

**Units**: Candidates that participated in the election.

####**We can't use the simple difference in observed group means (SDO) to estimate the causal ATE. In this section, we'll try inverse propensity weighting instead.**

```python
obs = all_data.copy()
```

```python
treatment = []
for x in all_data['Net_Contribution']:
  if x == 0:
    treatment.append(0)
  else:
    treatment.append(1)

treatment = pd.Series(treatment)
```

```python
# switch categorical columns to binary or something

### race type: 0 = 'Regular', 1 = 'Special'
obs['Race Type'] = obs['Race Type'].fillna(0).replace('Regular', 0).replace('Special', 1)

### general status: 0 = 'None', 1 = 'On the Ballot'
obs['General Status'] = obs['General Status'].fillna(0).replace('None', 0).replace('On the Ballot', 1)

### primary status: -1 = 'Lost', 1 = 'On the Ballot', 2 = 'Advanced',
obs['Primary Status'] = obs['General Status'].replace('Lost', -1).replace('On the Ballot', 1).replace('Advanced', 2)

### primary runoff status: -1 = 'Lost', 0 = 'None', 1 = 'On the Ballot', 2 = 'Advanced',
obs['Primary Runoff Status'] = obs['General Status'].replace('Lost', -1).replace('None', 0).replace('On the Ballot', 1).replace('Advanced', 2)

### candidate office: 0 = 'H', 1 = 'S'
obs['Cand_Office'] = obs['Race Type'].replace('H', 0).replace('S', 1)

### party affiliation: 0 = 'Rep', 1 = 'Dem'
obs['Party Affiliation'] = obs['Party Affiliation'].replace('Rep', 0).replace('Dem', 1)

### office type: 0 = 'Representative', 1 = 'Senator', 2 = 'Governor'
obs['Office Type'] = obs['Race Type'].replace('Representative', 0).replace('Senator', 1).replace('Governor', 2)

### assumes that Nan is zero.
obs['Cash_On_Hand_BOP'] = obs['Cash_On_Hand_BOP'].fillna(0)

### assumes that Nan is zero.
obs['Cash_On_Hand_COP'] = obs['Cash_On_Hand_BOP'].fillna(0)

### assumes that Nan is zero.
obs['Primary %'] = obs['Primary %'].fillna(0)
```

```python
# those who are not reported to win or lose, they are zero
outcome = all_data['Won Primary'].fillna(0).replace('No', -1).replace('Yes', 1)
```

```python
X = obs.drop(['Candidate', 'State', 'District', 'Race Primary Election Date', 
              'Cand_Office_St', 'Cand_Office_Dist', 'Cand_Party_Affiliation', 
              'Cand_Incumbent_Challenger_Open_Seat', 'Won Primary', 
              'Coverage_End_Date', 'Cand_Street_1', 'Cand_Street_2', 'Cand_City',
              'Cand_State', 'Cand_Zip', 'Coverage_Start_Date', 'Primary Status',
              'Debt_Owe_To_Committee', 'Debt_Owed_By_Committee', "Net_Contribution",
              "Individual_Itemized_Contribution", "Individual_Unitemized_Contribution",
              "Individual_Contribution", "Other_Committee_Contribution",
              "Party_Committee_Contribution", "Cand_Contribution", "Total_Contribution",
              ],
             axis=1)
```

```python
# imported from hw4
# fit a logistic regression model
from sklearn.linear_model import LogisticRegression as LR
lr = LR(penalty='none', max_iter=5000, random_state=0)
fitted = lr.fit(X, treatment)
```

```python
# showing the pscore computation
pscore = fitted.predict_proba(X)[:, 1]
pscore
```

```python
# creating a dataset with the pscores and treatment columns
obs_prop = all_data.copy()
obs_prop['treat'] = treatment
obs_prop['pscore'] = pscore
```

```python
# imported from hw4
def estimate_treatment_effect(lr, X, Y, Z):
  ex = lr.predict_proba(X)[:, 1]
  return np.mean(Z * Y / ex) - np.mean((1 - Z) * Y / (1 - ex))
```

```python
# calculating the ATE of net contribution of a candidate as we did in hw4
ate = estimate_treatment_effect(fitted, X, outcome, treatment)
print("Average Treatment Effect w/ LR: " + str(ate))
```

Computes the IPW estimate for the ATE as it was done in lab8.

```python
# imported from lab8
n = len(obs_prop)
e = obs_prop['pscore']
y = obs_prop['Net_Contribution']
treat = obs_prop['treat'] == 1
ipw_estimate = ( np.sum(y[treat] / e[treat])/n ) - ( np.sum(y[~treat] / (1-e[~treat]))/n ) # SOLUTION
print("Average Treatment Effect w/out LR: " + str(ipw_estimate))
```
